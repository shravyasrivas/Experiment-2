```python
#shravya srivastva
#aim: study of list

#LISTS---------
listfruits=["apple","orange","cherry"]
#duplicates are allowed in lists
#lists exits for numbers too
#append
listfruits.append("blue berry")
#printing 2
print(listfruits[-2],listfruits[0])
```

```python
#banana/2 aftr apple and not at the end
listfruits=["apple","orange","cherry"]
listfruits.insert(1,2)
print(listfruits)
#replacing 2 with banana
listfruits[1]="banana"
print(listfruits)
```

```python
#duplicated
listfruits=["apple","apple","banana"]
print(listfruits)
#to find number of elements
print(len(listfruits))
```

```python
#numbered list
listn=[1,2,3]
print(listn)
```

```python
#combining lists
list1=["apple","banana","cherry"]
list2=["blue berry","strawberry","cranberry"]
#by addition
print(list1+list2)
#by extend
list1.append("watermelon")
list1.extend(list2)
print(list1)
```

```python
list1=["apple","banana","cherry","kiwi","mango","watermelon","dragon fruit"]
#printing few numbers
print(list1[2:7])
#printing till end
print(list1[3:])
#from start
#ending always (n-1) element tak jayega
print(list1[:5])
print(list1[:-2])
print(list1[-3:])
```
